## First data Cleaning process: (8/11/2025)
### As we see up there, some the property name and host name value might have some additional review information that could be helpful for the process
### Hence: I want to fill the name missing value with no name and host name to unknown.
### Then I want to dropna to delete the missing value: because we know its all related to no review which wont will be useful!

```python
df['name'].fillna('No name', inplace=True)
df['host_name'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head(10)
```

```python
#Delete unnamed:0 column and reset index!
df.drop('Unnamed: 0', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
```

### This is the finishing line between first data cleaning, I want to use boxplot to identify the potential outlier also deleting duplicated name

```python
### Do descriptive statistic to see if its reasonable
df.describe()
```

From descriptive analysis, we know that min minimum_nights, number_of_reviews, reviews_per_month > 0 which sounds fine. Now we wanted to check an anomaly for availability, and price.

```python
#boxplot for price
sns.boxplot(df['price'])
plt.show()
```

```python
#it seems that there is an outlier in the price column. Hence, we will remove the quantile 0.95 to make it more reasonable in value
price_outlier = df['price'].quantile(0.95)
df.drop(df[df['price'] > price_outlier].index, inplace=True)
df.describe()
```

```python
#boxplot for price once again
sns.boxplot(df['price'])
plt.show()
```

```python
#generate boxplot for price for every room types
sns.boxplot(x='room_type', y='price', data=df)
plt.show()
```

```python
#find the 99% quantile of price for each room type
df.groupby('room_type')['price'].quantile(0.99)
```

```python
#To make it tidier, we remove the outlier for each room type by 99% quantile
thresholds = df.groupby('room_type')['price'].quantile(0.99)
df['threshold'] = df['room_type'].map(thresholds)
df = df[df['price'] <= df['threshold']]
df = df.drop(columns=['threshold'])
df.describe()
```

```python
sns.boxplot(x='room_type', y='price', data=df)
plt.show()
```

```python
#Check the amount of outlier in minimum_nights by boxplot
sns.boxplot(df['minimum_nights'])
plt.show()
```

```python
#delete outlier in minimum_nights by 95% quantile
minimum_nights_outlier = df['minimum_nights'].quantile(0.95)
df.drop(df[df['minimum_nights'] > minimum_nights_outlier].index, inplace=True)
df.describe()
```

```python
sns.boxplot(df['minimum_nights'])
plt.show()
```

```python
#Check an boxplot graph of availibility
sns.boxplot(df['availability_365'])
plt.show()
```

```python
#When the availability is 0, it would not be useful for the data
df.drop(df[df['availability_365'] == 0].index, inplace=True)
df.describe()
```

```python
#check the box plot of number of reviews
sns.boxplot(df['number_of_reviews'])
plt.show()
```

```python
#delete the outliers of number of reviews by 95% quantile
number_of_reviews_outlier = df['number_of_reviews'].quantile(0.95)
df.drop(df[df['number_of_reviews'] > number_of_reviews_outlier].index, inplace=True)
df.describe()
```

```python
sns.boxplot(df['number_of_reviews'])
plt.show()
```

### Duplications handling

```python
sns.boxplot(df['number_of_reviews_ltm'])
plt.show()
```

```python
#delete outliers of number_of_reviews_ltm by 99% quantile
number_of_reviews_ltm_outlier = df['number_of_reviews_ltm'].quantile(0.99)
df.drop(df[df['number_of_reviews_ltm'] > number_of_reviews_ltm_outlier].index, inplace=True)
df.describe()
```

```python
# There is duplicated name but id seems different. Lets see the duplicated name table to identify further
df[df.duplicated('name')]
```

After data wrangling, I saw there is difference in lat and long and it might have a SEO trick that put it in the same name. Double checking with lat long:

```python
duplicates = df[df.duplicated(subset=['name'], keep=False)]
plt.figure(figsize=(10,6))
plt.scatter(duplicates['longitude'], duplicates['latitude'], alpha=0.6)
plt.title('Listings with Duplicate Names')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

Conclusion: Do not delete the data because of the lat long validation shows slight of difference for each of them even though there is a denser ones
